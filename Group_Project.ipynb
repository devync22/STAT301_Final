```python
# Missing values in the data table
import pandas as pd

# Load the dataset in Python to perform wrangling and visualization
file_path = '/Users/lijiahao/stat301 Group Project/data/adult (3)/adult.data'
column_names = [
    "age", "workclass", "fnlwgt", "education", "education_num", 
    "marital_status", "occupation", "relationship", "race", 
    "sex", "capital_gain", "capital_loss", "hours_per_week", 
    "native_country", "income"
]
data = pd.read_csv(file_path, header=None, names=column_names)

# Strip whitespace from all text columns to clean up any trailing spaces
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Handle missing values by marking "?" as NaN
data = data.replace("?", pd.NA)

# Display a summary of missing values
missing_summary = data.isna().sum()
print(missing_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gender vs. Income Visualization
# Description:
# This visualization shows the proportion of individuals earning <=50K and >50K in each gender group.
# The stacked bar chart highlights the relative differences in income proportions between males and females,
# allowing for a clear comparison within each gender group.

# Justification:
# Proportional comparisons are essential here because raw counts can be misleading if the groups are imbalanced in size.
# By using proportions, we better understand income disparity while normalizing for gender group size.

plt.figure(figsize=(10, 6))

# Calculate proportions for gender
gender_income_prop = data.groupby(['sex', 'income']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

# Plot the proportions
gender_income_prop.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title('Proportion of Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'], rotation=0)
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()

# Education vs. Income Visualization
# Description:
# This visualization shows the proportion of individuals earning <=50K and >50K across different education levels.
# The horizontal stacked bar chart makes it easy to see how income distribution varies with education level.

# Justification:
# Education is a critical factor influencing income. Using a proportional horizontal bar chart allows us to
# visually compare categories while accounting for the varying sample sizes within education levels. 
# Sorting by frequency of education levels ensures the most common levels are displayed prominently.

plt.figure(figsize=(12, 8))

# Calculate proportions for education
education_income_prop = data.groupby(['education', 'income']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

# Sort by overall frequency of education
education_order = data['education'].value_counts().index

# Plot the proportions
education_income_prop.loc[education_order].plot(kind='barh', stacked=True, figsize=(12, 8), color=['skyblue', 'orange'])
plt.title('Proportion of Income by Education Level')
plt.xlabel('Proportion')
plt.ylabel('Education Level')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns



# Clean and preprocess the data
data['income'] = data['income'].apply(lambda x: 1 if x.strip() == '>50K' else 0)

# Encode categorical variables using one-hot encoding
categorical_vars = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country']
data_encoded = pd.get_dummies(data, columns=categorical_vars, drop_first=True)

# Split data into features and target
X = data_encoded.drop('income', axis=1)
y = data_encoded['income']

# Split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Correlation with target variable
correlations = data_encoded.corr()['income'].abs().sort_values(ascending=False)

# Select the top 3 predictors based on correlation
top_features = correlations[1:4].index.tolist()

print("Top 3 predictors selected based on correlation:", top_features)

# Train a logistic regression model using the top 3 features
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train[top_features], y_train)

# Evaluate the model
y_pred = logreg.predict(X_test[top_features])
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Display results
print(f"Model Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", report)

# Visualize feature importance
coefficients = pd.DataFrame({
    "Feature": top_features,
    "Coefficient": logreg.coef_[0]
}).sort_values(by="Coefficient", key=abs, ascending=False)

plt.barh(coefficients["Feature"], coefficients["Coefficient"])
plt.xlabel("Coefficient Value")
plt.ylabel("Feature")
plt.title("Feature Importance for Logistic Regression")
plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm

# Add the predictors and target variable
glm_data = data[["income", "education_num", "age", "marital_status"]]

# Encode the categorical variable `marital_status` using one-hot encoding
glm_data = pd.get_dummies(glm_data, columns=["marital_status"], drop_first=True)

# Rename columns to remove problematic characters (e.g., hyphens)
glm_data.rename(columns=lambda x: x.replace("-", "_"), inplace=True)

# Check the column names to confirm changes
print(glm_data.columns)

# Define the formula using the renamed columns
formula = "income ~ education_num + age + marital_status_Married_civ_spouse + marital_status_Never_married"

# Fit the GLM model
income_glm = sm.GLM.from_formula(formula=formula, data=glm_data, family=sm.families.Binomial()).fit()

# Display the summary of the model
print(income_glm.summary())

# Make predictions on the training data
glm_data["predicted"] = income_glm.predict(glm_data)
glm_data["predicted_binary"] = (glm_data["predicted"] > 0.5).astype(int)

# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report

accuracy_glm = accuracy_score(glm_data["income"], glm_data["predicted_binary"])
report_glm = classification_report(glm_data["income"], glm_data["predicted_binary"])

# Display results
print(f"Model Accuracy: {accuracy_glm:.2f}")
print("\nClassification Report:\n", report_glm)

# Visualize coefficients
coefficients_glm = pd.DataFrame({
    "Feature": income_glm.params.index,
    "Coefficient": income_glm.params.values
}).sort_values(by="Coefficient", key=abs, ascending=False)

import matplotlib.pyplot as plt

plt.barh(coefficients_glm["Feature"], coefficients_glm["Coefficient"])
plt.xlabel("Coefficient Value")
plt.ylabel("Feature")
plt.title("Feature Importance for Logistic Regression")
plt.tight_layout()
plt.show()
```

The logistic regression model achieved an accuracy of 82%, indicating good overall performance. The results show that being married (marital_status_Married_civ_spouse), higher education level (education_num), and age positively impact the likelihood of earning more than $50K, while being never married (marital_status_Never_married) negatively impacts it. The model struggles with recall for the minority class (income > $50K), indicating potential room for improvement through resampling or feature engineering. Overall, the results align with expectations and confirm that the selected predictors are significant in addressing the research question.
